```python
x = train_data['lotsize']
y = train_data['price']

plt.scatter(x,y, c='b')
plt.title('lotsize vs price')
plt.xlabel('lotsize')
plt.ylabel('price')
plt.show()
```

```python
# Function to calculate linear regression
def linear_reg():
    x_diff = []
    y_diff = []
    sum_of_diff = 0
    sum_of_x_diff_sqrd = 0
    
    # Get mean of x-axis and y-axis
    x_mean = x.mean()
    y_mean = y.mean()
    
    # Get differences between x-points and x_mean
    for val in x:
        x_diff.append( val - x_mean )
        
    # Get differences between y-points and y_mean
    for val in y:
        y_diff.append( val - y_mean )
        
    # Get product then sum of differences
    for i in range( len( x_diff ) ):
        sum_of_diff += x_diff[i] * y_diff[i]
        sum_of_x_diff_sqrd += x_diff[i] ** 2
        
    # Get gradient
    gradient = sum_of_diff / sum_of_x_diff_sqrd
    gradient = round( gradient , 3 )
    
    # Get constant
    constant = y_mean - ( gradient * x_mean )
    constant = round( constant , 3 )
    
    # Print linear regression model
    print("y = " + str(gradient) + "x + " + str(constant))
    
```

```python
linear_reg()
```

```python
x = housing['lotsize'].values.reshape(-1,1)
y = housing['price'].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
linearRegressor = LinearRegression()
linearRegressor.fit(x_train, y_train)
```

```python
y_pred = linearRegressor.predict(x_test)
```

```python
plt.scatter(x_train, y_train,  color='gray')
plt.plot(x_train, linearRegressor.predict(x_train) , color='blue', linewidth=2)
plt.title("Scatter PLot")
plt.xlabel("Lotsize")
plt.ylabel("Price")
plt.show()
```

```python
from sklearn import metrics

print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
```
