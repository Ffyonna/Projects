## Visualization
- Allows for looking at several data points together and see what they are telling us.

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
census_houses = pd.read_excel('census population households corrected.xlsx', thousands = ',')
census_houses.head()
```

```python
census_area = pd.read_excel('census area density corrected.xlsx', thousands = ',')
census_area.head()
```

```python
census_area.info()
```

```python
census_houses.info()
```

```python
census_houses.fillna(0, inplace = True)
census_houses.info()
```

```python
census_houses['Population'] = census_houses['Population'].astype(np.int64)
census_houses['Households'] = census_houses['Households'].astype(np.int64)

census_houses.info()
```

```python
census = pd.merge(census_houses, census_area, on = 'Sub County')
census.head()
```

```python
counties = census['County'][:20]
subs = census['Sub County'][:20]
population = census['Population'][:20]
households = census['Households'][:20]
size = census['Average Household Size'][:20]
area = census['Area'][:20]
density = census['Density'][:20]
```

### Basic plot

```python
plt.plot(subs, population)
```

### Figure size

```python
plt.figure(figsize = (10,7))

plt.plot(subs, population)
```

### Labels

```python
plt.figure(figsize = (10,6))

plt.xticks(rotation = 90)

plt.plot(subs, population)
```

### Bar plot

```python
plt.figure(figsize = (12,7))

plt.xticks(rotation = 75)

plt.xlabel('Sub Counties', fontsize = 15)
plt.ylabel('Population', fontsize = 15)
plt.title('Sub County vs Population', fontsize = 15)

plt.bar(subs, population)
```

```python
plt.figure(figsize = (12,7))

plt.xticks(rotation = 75)

plt.xlabel('Sub Counties', fontsize = 15)
plt.ylabel('Population', fontsize = 15)
plt.title('Sub County vs Population', fontsize = 15)

plt.bar(subs, population, width = 0.4, color = 'purple', edgecolor = 'black')
```

### Line plots

```python
plt.figure(figsize = (12,7))

plt.xticks(rotation = 75)

plt.plot(subs, population)
plt.plot(subs, households)
```

```python
plt.figure(figsize = (12,7))

plt.xticks(rotation = 75)

plt.plot(subs, population, 'bo')
plt.plot(subs, households, 'r+')
```

```python
plt.figure(figsize = (12,7))

plt.xticks(rotation = 75)

plt.plot(subs, population, color = 'purple', marker = 'x', linestyle = 'dashed') #--
plt.plot(subs, households, color = 'green', marker = '^', ls = 'dotted') #:
plt.plot(subs, size, color = 'purple', alpha = .6, marker = 'o', ls = 'dashdot') #-.
```

```python
plt.figure(figsize = (12,7))

plt.xticks(rotation = 75)

plt.plot(subs, population, color = 'purple', marker = 'x', linestyle = 'dashed', label = 'Population') #--
plt.plot(subs, households, color = 'green', marker = '^', ls = 'dotted', label = 'Households') #:
plt.plot(subs, size, color = 'purple', alpha = .6, marker = 'o', ls = 'dashdot', label = 'Average Households Size') #-.

plt.legend()
```

### Scatter plots

```python
plt.figure(figsize = (12,6))

plt.xticks(rotation = 75)

plt.scatter(subs, population)
```

```python
plt.figure(figsize = (12,6))

plt.xticks(rotation = 75)

plt.scatter(subs, population, s = area, c  = 'purple')
```

### Multiple plots

```python
plt.figure(figsize = (15,7))

plt.subplot(2,2,1)
plt.plot(subs, population)
plt.xticks(rotation = 90)
plt.title('Population')

plt.subplot(2,2,2)
plt.plot(subs, area)
plt.xticks(rotation = 90)
plt.title('Area')

plt.suptitle('Population and Area')
```

```python
fig, subplot = plt.subplots(1, 2, figsize = (12,7))

subplot[0].plot(subs, population)


subplot[1].plot(subs, area)

plt.suptitle('Populatuon and Area')
```

### Seaborn

```python
import seaborn as sns
```

```python
plt.figure(figsize = (14,6))

plt.xticks(rotation = 30)

sns.barplot(subs, population)
```

```python
plt.figure(figsize = (14,6))

plt.xticks(rotation = 30)

sns.barplot(subs, population, hue = counties)
```

```python
sns.relplot(x = 'Sub County', y = 'Population', data = census.head(20), height = 6, aspect = 2, kind = 'scatter')

plt.xticks(rotation = 75)
```

```python
sns.relplot(x = 'Sub County', y = 'Population', data = census.head(10), height = 5, aspect = 2, \
           col = 'County')

plt.xticks(rotation = 75)
```

```python
sns.relplot(x = 'Sub County', y = 'Area', data = census.head(20), kind = 'line', height = 5, aspect = 2)

plt.xticks(rotation = 75)
```

```python
plt.figure(figsize = (10,6))

sns.distplot(census['Population'], kde = False, bins = 40)
```
